```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
```

```python
data = pd.read_csv('/kaggle/input/medicine-recommendation-system-dataset/Training.csv', delimiter=',')
data
```

```python
data = data.drop_duplicates()
data
```

```python
categorical_cols = data.columns
categorical_cols = categorical_cols[:-1]
target = ['prognosis']

len(categorical_cols), len(target)
```

```python
X = data[categorical_cols]
y = data[target]
```

```python
model = DecisionTreeClassifier(random_state=42)

model.fit(X, y)
```

```python
classes = np.unique(y)

tree = export_graphviz(model, 
                       out_file=None,
                       feature_names=categorical_cols,
                       class_names=classes,
                       filled=True,
                       rounded=True,
                       special_characters=True)

graph = graphviz.Source(tree)
graph.format = 'png'
graph.render('/kaggle/working/single_tree', cleanup=True)
```
