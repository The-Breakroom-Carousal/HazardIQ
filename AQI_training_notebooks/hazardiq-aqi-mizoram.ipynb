```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
aizawl_meteo = pd.read_csv('/kaggle/input/mizoram-aqi/Mizoram/Metrological Data/Aizawl.csv', delimiter=',')
aizawl_meteo
```

```python
aizawl_pollutants = pd.read_csv('/kaggle/input/mizoram-aqi/Mizoram/Pollutants/Aizawl.csv', delimiter= ',')
aizawl_pollutants['Timestamp'] = aizawl_meteo['time']
aizawl_pollutants.rename(columns={'Timestamp': 'time'}, inplace=True)
aizawl_pollutants
```

```python
aizawl = pd.merge(aizawl_meteo.iloc[:, :], aizawl_pollutants.iloc[:, 0:3], on='time')
aizawl
```

```python
aizawl['city'] = 'Aizawl'
aizawl
```

```python
aizawl.info()
```

```python
aizawl = aizawl.interpolate()
aizawl.info()
```

```python
for col in ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']:
    delay1_col = col + "_delay1"
    aizawl[delay1_col] = aizawl[col].shift(1)
    delay2_col = col + "_delay2"
    aizawl[delay2_col] = aizawl[col].shift(2)
    aizawl.loc[0, delay1_col] = aizawl[col].iloc[-1]
    aizawl.loc[1, delay2_col] = aizawl[col].iloc[-1]
    aizawl.loc[0, delay2_col] = aizawl[col].iloc[-2]

aizawl
```

```python
mizoram_data = aizawl
mizoram_data
```

```python
mizoram_data['city'] = mizoram_data['city'].astype('category')
```

```python
input_cols, target_cols = ['temperature_2m (°C)', 'relative_humidity_2m (%)', 'rain (mm)',
       'surface_pressure (hPa)', 'wind_speed_10m (km/h)',
       'wind_speed_100m (km/h)', 'wind_direction_10m (°)',
       'wind_direction_100m (°)', 'city', 'PM2.5 (µg/m³)_delay1', 'PM2.5 (µg/m³)_delay2', 'PM10 (µg/m³)_delay1',
       'PM10 (µg/m³)_delay2'], ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']
```

```python
X, y = mizoram_data[input_cols], mizoram_data[target_cols]
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(
    tree_method='gpu_hist',
    n_estimators=50,
    learning_rate=0.1,
    min_child_weight=5,
    enable_categorical=True,
)
```

```python
model.fit(X, y['PM2.5 (µg/m³)'], verbose = True)
```

```python
pred_check = model.predict(X)
```

```python
plt.plot(y['PM2.5 (µg/m³)'])
plt.plot(pred_check)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y['PM2.5 (µg/m³)'], pred_check)
print("R² score:", r2)
```

```python
import joblib
joblib.dump(model, 'Mizoram_PM2.5_µgm³.pkl')
```

```python
model.fit(X, y['PM10 (µg/m³)'])
```

```python
pred_check2 = model.predict(X)
```

```python
r2 = r2_score(y['PM10 (µg/m³)'], pred_check2)
print("R² score:", r2)
```

```python
plt.plot(y['PM10 (µg/m³)'])
plt.plot(pred_check2)
plt.show()
```
