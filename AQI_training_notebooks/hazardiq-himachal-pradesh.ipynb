```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
baddi_meteo = pd.read_csv('/kaggle/input/himachal-pradesh-aqi/Himachal Pradesh/Metrological Data/Baddi.csv', delimiter=',')
baddi_meteo
```

```python
baddi_pollutants = pd.read_csv('/kaggle/input/himachal-pradesh-aqi/Himachal Pradesh/Pollutants/Baddi.csv', delimiter= ',')
baddi_pollutants['Timestamp'] = baddi_meteo['time']
baddi_pollutants.rename(columns={'Timestamp': 'time'}, inplace=True)
baddi_pollutants
```

```python
baddi = pd.merge(baddi_meteo.iloc[:, :], baddi_pollutants.iloc[:, 0:3], on='time')
baddi
```

```python
baddi['city'] = 'Baddi'
baddi
```

```python
baddi.info()
```

```python
baddi = baddi.interpolate()
baddi.info()
```

```python
for col in ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']:
    delay1_col = col + "_delay1"
    baddi[delay1_col] = baddi[col].shift(1)
    delay2_col = col + "_delay2"
    baddi[delay2_col] = baddi[col].shift(2)
    baddi.loc[0, delay1_col] = baddi[col].iloc[-1]
    baddi.loc[1, delay2_col] = baddi[col].iloc[-1]
    baddi.loc[0, delay2_col] = baddi[col].iloc[-2]

baddi
```

```python
himachalpradesh_data = baddi
```

```python
himachalpradesh_data['city'] = himachalpradesh_data['city'].astype('category')
```

```python
input_cols, target_cols = ['temperature_2m (°C)', 'relative_humidity_2m (%)', 'rain (mm)',
       'surface_pressure (hPa)', 'wind_speed_10m (km/h)',
       'wind_speed_100m (km/h)', 'wind_direction_10m (°)',
       'wind_direction_100m (°)', 'city', 'PM2.5 (µg/m³)_delay1', 'PM2.5 (µg/m³)_delay2', 'PM10 (µg/m³)_delay1',
       'PM10 (µg/m³)_delay2'], ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']
```

```python
X, y = himachalpradesh_data[input_cols], himachalpradesh_data[target_cols]
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(
    tree_method='gpu_hist',
    n_estimators=150,
    learning_rate=0.1,
    min_child_weight=5,
    enable_categorical=True,
)
```

```python
model.fit(X, y['PM2.5 (µg/m³)'], verbose = True)
```

```python
pred_check = model.predict(X)
```

```python
plt.plot(y['PM2.5 (µg/m³)'])
plt.plot(pred_check)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y['PM2.5 (µg/m³)'], pred_check)
print("R² score:", r2)
```

```python
import joblib
joblib.dump(model, 'HimachalPradesh_PM2.5_µgm³.pkl')
```

```python
model.fit(X, y['PM10 (µg/m³)'])
```

```python
pred_check2 = model.predict(X)
```

```python
r2 = r2_score(y['PM10 (µg/m³)'], pred_check2)
print("R² score:", r2)
```

```python
plt.plot(y['PM10 (µg/m³)'])
plt.plot(pred_check2)
plt.show()
```
