```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
ambala_meteo = pd.read_csv('/kaggle/input/haryana-aqi/Haryana/Metrological Data/Ambala.csv', skiprows = 2, delimiter=',')
ambala_meteo
```

```python
ambala_pollutants = pd.read_csv('/kaggle/input/haryana-aqi/Haryana/Pollutants/Ambala.csv', delimiter= ',')
ambala_pollutants['Timestamp'] = ambala_meteo['time']
ambala_pollutants.rename(columns={'Timestamp': 'time'}, inplace=True)
ambala_pollutants
```

```python
ambala = pd.merge(ambala_meteo.iloc[:, :], ambala_pollutants.iloc[:, 0:3], on='time')
ambala
```

```python
ambala['city'] = 'Ambala'
ambala
```

```python
ambala.info()
```

```python
ambala = ambala.interpolate()
ambala.info()
```

```python
for col in ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']:
    delay1_col = col + "_delay1"
    ambala[delay1_col] = ambala[col].shift(1)
    delay2_col = col + "_delay2"
    ambala[delay2_col] = ambala[col].shift(2)
    ambala.loc[0, delay1_col] = ambala[col].iloc[-1]
    ambala.loc[1, delay2_col] = ambala[col].iloc[-1]
    ambala.loc[0, delay2_col] = ambala[col].iloc[-2]

ambala
```

```python
haryana_data = pd.concat([ambala, bahadurgarh, ballabgarh, bhiwani, faridabad, fatehabad, gurugram, panipat, sirsa, sonipat], axis=0, ignore_index=True)
haryana_data
```

```python
haryana_data['city'] = haryana_data['city'].astype('category')
```

```python
input_cols, target_cols = ['temperature_2m (°C)', 'relative_humidity_2m (%)', 'rain (mm)',
       'surface_pressure (hPa)', 'wind_speed_10m (km/h)',
       'wind_speed_100m (km/h)', 'wind_direction_10m (°)',
       'wind_direction_100m (°)', 'city', 'PM2.5 (µg/m³)_delay1', 'PM2.5 (µg/m³)_delay2', 'PM10 (µg/m³)_delay1',
       'PM10 (µg/m³)_delay2'], ['PM2.5 (µg/m³)', 'PM10 (µg/m³)']
```

```python
X, y = haryana_data[input_cols], haryana_data[target_cols]
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(
    tree_method='gpu_hist',
    n_estimators=150,
    learning_rate=0.1,
    min_child_weight=5,
    enable_categorical=True,
)
```

```python
model.fit(X, y['PM2.5 (µg/m³)'], verbose = True)
```

```python
pred_check = model.predict(X)
```

```python
plt.plot(y['PM2.5 (µg/m³)'])
plt.plot(pred_check)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y['PM2.5 (µg/m³)'], pred_check)
print("R² score:", r2)
```

```python
import joblib
joblib.dump(model, 'Haryana_PM2.5_µgm³.pkl')
```

```python
model.fit(X, y['PM10 (µg/m³)'])
```

```python
pred_check2 = model.predict(X)
```

```python
r2 = r2_score(y['PM10 (µg/m³)'], pred_check2)
print("R² score:", r2)
```
